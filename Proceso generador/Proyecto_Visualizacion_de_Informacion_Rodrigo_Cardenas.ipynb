```python
import pandas as pd
path = "/content/drive/MyDrive/Colab Notebooks/Visualizacion de Informacion y Analisis Visual/Proyecto/"
```

```python
df = pd.read_csv(path+"steam.csv")
df
```

```python
len(df.genres.unique())
```

```python
df["genresLista"] = df.genres.apply(lambda x: x.split(";"))
generos = []
for i,row in df.iterrows():
  for g in row.genresLista:
    if g not in generos:
      generos.append(g)
print(generos)
print(len(generos))
```

```python
generosFinal = ['Action', 'Strategy', 'Adventure', 'Indie', 'RPG', 'Simulation', 'Racing', 'Massively Multiplayer', 'Sports']
len(generosFinal)
```

```python
df["genero"] = "Todos"
```

```python
dfFinal = df

for g in generosFinal:
  auxlistaIndices = []
  for i in df.genresLista:
    auxlistaIndices.append(g in i)
  dfAux = df[auxlistaIndices].copy()
  dfAux["genero"] = g 
  dfFinal = dfFinal.append(dfAux)

print(len(df))
print(len(dfFinal))
dfFinal.groupby(["genero"]).count()["appid"]
```

```python
dfFinal["positiveReviewsPct"] = dfFinal.positive_ratings / (dfFinal.positive_ratings + dfFinal.negative_ratings)
dfFinal["positiveReviewsPct"]
```

```python
dfFinal
```

```python
dfFinal["year"] = dfFinal.release_date.apply(lambda x: x[:4])
dfFinal["year"]
```

```python
import altair as alt
```

```python
import math
math.log(2)
```

```python
dfScatter5000 = dfFinal[(dfFinal.genero != "Todos") &  (dfFinal.average_playtime > 1)] \
  .sort_values(by=["genero","positive_ratings"], inplace=False, ascending=False)\
  .groupby(["genero"]).head(500)
dfScatter5000["average_playtime_log"] = dfScatter5000.average_playtime.map(math.log)
```

```python
dfScatter5000.to_csv("datosComparadorScatter.csv",sep=";",index=False)
```

```python
ordenOwners = ['0-20000',
'20000-50000',
'50000-100000',
'100000-200000',
'200000-500000',
'500000-1000000',
'1000000-2000000',
'2000000-5000000',
'5000000-10000000',
'10000000-20000000',
'20000000-50000000',
'50000000-100000000',
'100000000-200000000',
]
```

```python
brush = alt.selection_interval(
    encodings=['x'] # limit selection to x-axis (year) values
)

input_dropdown = alt.binding_select(options=[None] + generosFinal, labels = ["All"] + generosFinal, name='Genre ')
selection = alt.selection_single(fields=['genero'], bind=input_dropdown)

input_dropdownOwner = alt.binding_select(options=[None] + ordenOwners, labels = ["All"] + ordenOwners, name='Owners ')
selectionOwner = alt.selection_single(fields=['owners'], bind=input_dropdownOwner)


selectionOwnerLegend = alt.selection_multi(fields=['owners'], bind='legend')
# dynamic query histogram
years = alt.Chart(dfScatter5000).mark_bar().add_selection(
    brush
).encode(
    alt.X('release_date:T', title='Video Games by Release Year'),
    alt.Y('count():Q', title=None)
).properties(
    width=650,
    height=50
)

# dynamic query histogram
brushPrecio = alt.selection_interval(
    encodings=['x'] # limit selection to x-axis (year) values
)

price = alt.Chart(dfScatter5000).mark_bar().add_selection(
    brushPrecio
).encode(
    alt.X('price:Q', title='Video Games by Price' ),#,scale=alt.Scale(domain=[0,100]) ),
    alt.Y('count():Q', title=None)
).properties(
    width=650,
    height=50
)

#slider = alt.binding_range(min=0, max=1000, step=1, name='Rango Precio:')
#selectorPrecio = alt.selection_single(name="price", fields=['price'],
#                                bind=slider, init={'price': 100})

# scatter plot, modify opacity based on selection
ratings = alt.Chart(dfScatter5000).mark_circle(size = 80).encode(
    x=alt.X('positiveReviewsPct:Q',title = "Positive Reviews Percentage"),
    y=alt.Y('average_playtime_log:Q', title = "Log of Average Played Time"),
    tooltip=['name:N','release_date:T','developer:N','publisher:N','owners:N','categories:N','genres:N','positive_ratings:Q','negative_ratings:Q','average_playtime:Q', 'price:Q']
    ,color = alt.Color('owners:O', scale=alt.Scale(domain=ordenOwners, scheme='purples'))
    ,opacity=alt.condition(selectionOwnerLegend, alt.value(0.75), alt.value(0.01))
).properties(
    width=650,
    height=400,
).add_selection(
    selection 
).add_selection(
    selectionOwner 
).add_selection(
    selectionOwnerLegend
).transform_filter(
    selection 
).transform_filter(
    selectionOwner 
).transform_filter(
    brush
).transform_filter(
    brushPrecio
).interactive()



alt.vconcat(years, ratings, price ).properties(spacing=5, padding  = 5)
```
